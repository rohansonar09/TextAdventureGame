```python
import time
import random
name = input (prompt = "What's your name?") #Storing the user's name

#Intro
print ("\nHello " +  name +", We use numbers quite often to describe various things - weights, heights, speeds, populations and so on.")
print("Today I've set a unique challenge to address your intution about numbers.")
print("\nThis game is loosely based on Danish game show 'Gimme 5' and its Indian version 'Kam ya zyada'.")
print("It consists of 3 stages. In the first stage, you have to answer only as “More” or “Less”.")
print("The 2nd stage will test your general knowledge where you have to answer by entering correct option (either a, b or c). ")
print("The third and last stage will test your estimation skills where you have to pick a correct option in the range of 10.")
print("\nRemember, you must answer atleast 1 question correctly in Stage 1 and 2 in order to qualify for stage 3.")
print("\nAlso, any invalid entry will result in restarting the game.")

time.sleep(10)


#Handle failure condition
def failGame():
    print("\n\nYou've failed to correctly answer a single question so your game ends here, but thanks for playing!")
    
#Game starts here
def playGame():
    print ("\nLet's get started with stage 1. Remember, the following questions needs to be answered only as 'More' or 'Less'.")
    #Storing the correct answers
    correct = 0
    time.sleep(4)
    try:
        more = ["More", "more", "MORE"]
        less = ["Less", "less", "LESS"]
        
        #Question for stage 1
        print ("\nThe height of leaning tower of Pisa is 58.5 meteres.")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in less or str(choice).lower() in more:
            if str(choice).lower() in less:
                correct += 1 #If correct, the user gets one point
            else:
                correct += 0
        else:
            print("Invalid Entry, please start the game again")
            playGame()
        
        #Question for stage 1
        print ("\nDistance from the Sun to Neptune is 2.98 billion miles")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in less or str(choice).lower() in more:
            if str(choice).lower() in less:
                correct += 1 #If correct, the user gets one point
            else:
                correct += 0
        else:
            print("Invalid Entry, please start the game again")
            playGame()

        #Question for stage 1
        print ("\nPercentage of Oxygen in atmosphere is 19%")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in less or str(choice).lower() in more:
            if str(choice).lower() in more:
                correct += 1 #If correct, the user gets one point
            else:
                correct += 0
        else:
            print("Invalid Entry, please start the game again")
            playGame()
    
        time.sleep(2)
        
        #Exercise using Random nos starts here
        print ("\n\nHere's a quick brain exercise for you before Stage 2, Answer a simple mathematical question and claim a bonus point. You have 2 chances for this !!")
        my_Answer = 1
        correct_1 = 0
        guess = 0

        opt_1 = random.randint(500,1000)
        opt_2 = random.randint(500,1000) 
        my_Answer = opt_1 + opt_2
        while guess != my_Answer:
            question="What is the sum of " + str(opt_1) +"+"+ str(opt_2)+"? "
            guess=float(input(prompt = question))
            if guess == my_Answer:
                print("Correct, Congratulations!! You earned 1 extra bonus point")
                correct_1 += 1
                correct += 1
            else:
                print ("Wrong Answer. You have one last chance")
                print("The correct answer was " +  str(my_Answer) +" ")
                correct_1 = correct_1 + 0
                opt_1 = random.randint(500,1000)
                opt_2 = random.randint(500,1000) 
                my_Answer = opt_1 + opt_2
                question="What is the sum of " + str(opt_1) +"+"+ str(opt_2)+"? "
                guess=float(input(prompt = question))
                if guess == my_Answer:
                    print("Correct, Congratulations!! You earned 1 extra bonus point")
                    correct_1 += 1
                    correct += 1
                else:
                    print ("Wrong Answer.")
                    break
        if correct_1 == 0:
            print("You lost the chance to score 1 bonus point, Better luck next time!!")
            print("The correct answer was " +  str(my_Answer) +" ")
        guess = 0
        #Exercise ends here
        
        time.sleep(2)
        
        #Stage 2 starts here
        print("\nWelcome to stage 2. Please enter the correct option from a,b or c.")

        time.sleep(4)
        
        #Question for stage 2
        print("\nWhich of these is the biggest? \n a. Width of football goal \n b. Distance a kangaroo can jump \n c. Estimated length of T-rex dinosaur \n")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in ('a','b','c'):
            if str(choice).lower() == 'b':
                correct += 1
            else:
                correct += 0 
        else:
            print("Invalid Entry, please start the game again")
            playGame()
        
        #Question for stage 2
        print("\nWhich of these is the greatest distance? \n a. Length of the Nile river \n b. Distance from London to New York \n c. Length of US/Canada border \n")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in ('a','b','c'):
            if str(choice).lower() == 'c':
                correct += 1
            else:
                correct += 0  
        else:
            print("Invalid Entry, please start the game again")
            playGame()
        
        #Question for stage 2
        print("\nWhich of these is the most numerous? \n a. Number of bones in the human body \n b. Number of buildings in New York City higher than 150 metres \n c. Number of aircraft carriers that have been constructed \n")
        choice = input(prompt = ">>> ")
        
        #Checking correctness of user input
        if str(choice).lower() in ('a','b','c'):
            if str(choice).lower() == 'b':
                correct += 1
            else:
                correct += 0 
        else:
            print("Invalid Entry, please start the game again")
            playGame()
            
        time.sleep(2)
        
        #Checking the failure criteria
        if correct == 0:
            failGame()
        else:
            #Stage 3 starts here
            print("\nWelcome to stage 3. Your answer is hidden in the range of 10. Please enter the correct option from a,b,c or d.")

            time.sleep(4)
            
            #Question for stage 3
            print("\nAccording to recent survey, what percentage of American adults consume coffee every day? \n a. 45-55 % \n b. 75-85 % \n c. 60-70 % \n d. 30-40 % \n")
            choice = input(prompt = ">>> ")
            
            #Checking correctness of user input
            if str(choice).lower() in ('a','b','c','d'):
                if str(choice).lower() == 'c':
                    correct += 1
                else:
                    correct += 0
            else:
                print("Invalid Entry, please start the game again")
                playGame()
            
            #Question for stage 3
            print("\nWhat percentage of Americans get the news from YouTube? \n a. 25-35 % \n b. 40-50 % \n c. 10-20 % \n d. 5-15 % \n")
            choice = input(prompt = ">>> ")
            
            #Checking correctness of user input
            if str(choice).lower() in ('a','b','c','d'):
                if str(choice).lower() == 'a':
                    correct += 1
                else:
                    correct += 0
            else:
                print("Invalid Entry, please start the game again")
                playGame()

            time.sleep(2)
            
            #Printing result
            if correct == 9:
                print("\nWoohoo, You're finished, " + name +". You got all answers correct including the bonus exercise, Excellent Job!!, Thanks for your participation! \n\n")
            elif correct >= 4:
                print("\nYou're finished, " + name +". You got", correct, "out of 9 correct including the bonus exercise, Great Job!! Thanks for your participation! \n\n")
            else:
                print("\nYou're finished, " + name +". You got", correct, "out of 9 correct including the bonus exercise. Thanks for your participation! \n\n")
                
            time.sleep(4)
            
            #Printing answers
            answers = [[1,'less'],[2,'less'],[3,'more'],[4,'b'],[5,'c'],[6,'b'],[7,'c'],[8,'a']]
            print("Correct answers are as follows - \n")
            for item in answers:
                print(item[0],item[1])
    except ValueError:
            print("Something went wrong. Please try again")
            playGame()
    return
            
playGame()
```
